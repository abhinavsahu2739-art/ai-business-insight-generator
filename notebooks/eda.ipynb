```python
#EDA STARTS
```

```python
# Cell1: KPI Metrics
print("Total Sales   :", round(df_clean["Sales"].sum(), 2))
print("Total Profit  :", round(df_clean["Profit"].sum(), 2))
print("Total Orders  :", len(df_clean))
print("Average Sales :", round(df_clean["Sales"].mean(), 2))
```

```python
#Cell2: Sales  by Category
category_sales = (
    df_clean.groupby("Category")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

category_sales
```

```python
#Cell 3: Profit by Category

category_profit =(
    df_clean.groupby("Category")["Profit"]
    .sum()
    .sort_values(ascending=False)
    )

category_profit
```

```python
#Cell 4: Sales by Region
region_sales = (
    df_clean.groupby("Region")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

region_sales
```

```python
#Cell 5: Top 10 States by Sales
top_states = (
    df_clean.groupby("State")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_states
```

```python
#1. Profit by Region
profit_region = (
    df_clean.groupby("Region")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

profit_region
```

```python
#Sales by Segment
segment_sales = (
    df_clean.groupby("Segment")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

segment_sales
```

```python
#Profit by Segment
segment_profit = (
    df_clean.groupby("Segment")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

segment_profit
```

```python
#Sales by Sub-Category
subcategory_profit = (
    df_clean.groupby("Sub-Category")["Profit"]
    .sum()
    .sort_values()
)

subcategory_profit
```
